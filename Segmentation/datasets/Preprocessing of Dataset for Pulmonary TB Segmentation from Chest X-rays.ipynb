```python
import os
import numpy as np
from tqdm.notebook import tqdm
from keras.preprocessing.image import load_img, img_to_array, save_img

os.chdir('/content/drive/My Drive/GitHub Repositories/State of Art Segmentation Network Architectures')
imagesDir_500 = './datasets/Pulmonary TB Segmentation from Chest X-rays Dataset/images(500*500)'
masksDir_500 = './datasets/Pulmonary TB Segmentation from Chest X-rays Dataset/masks(500*500)'

imagesDir_512 = './datasets/Pulmonary TB Segmentation from Chest X-rays Dataset/images(512*512)'
masksDir_512 = './datasets/Pulmonary TB Segmentation from Chest X-rays Dataset/masks(512*512)'

if not os.path.exists(imagesDir_512):
  os.makedirs(imagesDir_512)
if not os.path.exists(masksDir_512):
  os.makedirs(masksDir_512)

imageNames = sorted(os.listdir(imagesDir_500))
maskNames = sorted(os.listdir(masksDir_500))

# Pick an image, original size of image is 500*500, padd the edges of this image to make it 512*512  
for imageName in tqdm(imageNames):
  image = load_img(os.path.join(imagesDir, imageName), color_mode='grayscale', target_size=None)
  image_array = img_to_array(image, dtype='float32')
  padded_image_array = np.pad(array=image_array, pad_width=(6, 6), mode='edge')
  padded_image_array = np.expand_dims(padded_image_array[:, :, 0], axis=-1)
  save_img(os.path.join(imagesDir_512, imageName), padded_image_array)

# Pick a mask, original size of mask is 500*500, padd the edges of this mask to make it 512*512  
# Masks
for maskName in tqdm(maskNames):
  mask = load_img(os.path.join(masksDir, maskName), color_mode='grayscale', target_size=None)
  mask_array = img_to_array(mask, dtype='float32')
  padded_mask_array = np.pad(array=mask_array[:, :, 0], pad_width=(6, 6), mode='edge')
  padded_mask_array = np.expand_dims(padded_mask_array, axis=-1)
  save_img(os.path.join(masksDir_512, maskName), padded_mask_array)
```

```python
imageNames = sorted(os.listdir(imagesDir_512))
maskNames = sorted(os.listdir(masksDir_512))

IMG_SIZE = 512
images = []
masks = []
for imageName in tqdm(imageNames):
  image = load_img(os.path.join(imagesDir, imageName), color_mode='grayscale', target_size=(IMG_SIZE, IMG_SIZE), interpolation='lanczos')
  image_array =  img_to_array(image, data_format='channels_last', dtype='float32')
  images.append(image_array)

  mask = np.asarray(load_img(os.path.join(masksDir, imageName), color_mode='grayscale', target_size=(IMG_SIZE, IMG_SIZE), interpolation='lanczos'))
  mask_array =  img_to_array(mask, data_format='channels_last', dtype='float32')
  masks.append(mask_array)

images = np.asarray(images).reshape(-1, IMG_SIZE, IMG_SIZE, 1)
masks = np.asarray(masks).reshape(-1, IMG_SIZE, IMG_SIZE, 1)

np.save('./datasets/Pulmonary TB Segmentation from Chest X-rays Dataset/images.npy', images)
np.save('./datasets/Pulmonary TB Segmentation from Chest X-rays Dataset/masks.npy', masks)

print('Images and Masks are saved to Numpy Arrays in respective Directories')
```
